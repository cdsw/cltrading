```python
from scipy.signal import find_peaks
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic signal with closely spaced peaks
np.random.seed(0)
x = np.linspace(0, 10 * np.pi, 1000)
signal = np.sin(x) + np.random.normal(0, 0.3, size=len(x))

# Add closely spaced peaks
signal[300:310] += 1
signal[500:510] += 1.2
signal[700:710] += 1.5

# Find peaks with different distance parameters
distance_values = [10, 20, 50]  # Different distance values to observe peak detection

plt.figure(figsize=(10, 6))

for i, distance in enumerate(distance_values, 1):
    peaks, _ = find_peaks(signal, distance=distance)
    plt.subplot(3, 1, i)
    plt.plot(x, signal, label='Signal')
    plt.plot(x[peaks], signal[peaks], 'x', label=f'Peaks (distance={distance})', color='red')
    plt.legend()

plt.tight_layout()
plt.show()
```
